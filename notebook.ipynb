### Part 3: Analysis:

##### 1. The proportion of diamonds are between 0.30 and 1.08 carats

```python
carats = df['carat']

carats_list_succeeded = []
for x in carats:
    if x >= 0.30 and x <= 1.08:
        carats_list_succeeded.append(x)


# (X / Y) * 100
proportion = (len(carats_list_succeeded) / len(carats)) * 100
proportion = "{:.2f}".format(proportion)

print("Proportion of diamonds between 0.30 and 1.08 carats is: " + proportion + "%")
```

##### 2. Number of diamonds with x and y coordinates equal


```python
X = df['x']
Y = df['y']

xy_number_list = []

for x, y in zip(X, Y):
    if x == y:
        xy_number_list.append(x)

solution = len(xy_number_list)

print("Number of diamonds with x and y coordinates equal is: " + str(solution))
```

##### 3. Total number of diamonds with carats value less than mean carat value

```python
carats = df['carat']
carats_mean = carats.mean()

carats_scceded_list = []

for x in carats:
    if x < carats_mean:
        carats_scceded_list.append(x)


total_carats_less_than_mean = len(carats_scceded_list)
print("Total number of diamonds with carats value less than mean carat value is: " + str(total_carats_less_than_mean))
        
```

##### 4. Diamonds have a Premium cut or better ("Ideal") Note that the possible values for the quality of a cut are ranked in ascending order as follows: Fair / Good / Very Good / Premium / Ideal

```python
cut = df['cut']
cut_list = []

for x in cut:
    if x == "Ideal" or x == "Premium":
        cut_list.append(x)

number_of_dimonds_successed = str(len(cut_list))
print("Total number of diamonds with cut value Ideal or Premium is: " + number_of_dimonds_successed)
```

#### The highest price per carat and is its value

```python
carat = df['carat']
price = df['price']



carat_price_list = []
for x, y in zip(carat, price):
    carat_price_list.append([x, y])


max_carat_price = max(carat_price_list, key=lambda x: x[1])

max_carat_price
print("Max carat price is: " + str(max_carat_price[1]) + " And carat value is: " + str(max_carat_price[0]))
```

##### 6. First Using Seaborn, make boxplots of the diamond price for each distinct cut value and discuss any conclusions you can draw from the appearance of the boxplots.


```python
price = df['price']
cut = df['cut']

price_cut_list = []
for x, y in zip(price, cut):
    price_cut_list.append([x, y])

price_cut_list = pd.DataFrame(price_cut_list, columns=['Price', 'Cut'])

sns.boxplot(x=cut, y=price, data=pd.melt(price_cut_list)).set(title='Diamond price for each distinct cut value', xlabel='Cut', ylabel='Price')
```
